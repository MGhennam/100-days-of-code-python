# Option 2 : Directement depuis une URL brute


```python
url = "https://raw.githubusercontent.com/uiuc-cse/data-fa14/gh-pages/data/iris.csv"
df = pd.read_csv(url)
```

```python
df.head()
df.info()
df.describe()
```

```python
print(df.columns)
```

# Nuage de points (Scatter Plot)


```python
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x="sepal_length", y="sepal_width", hue="species")
plt.title("Relation entre sepal_length et sepal_width")
plt.show()
```

# Diagramme en boîte (Boxplot)

Utilisé pour visualiser la distribution et détecter les valeurs aberrantes.


```python
plt.figure(figsize=(8,6))
sns.boxplot(x="species", y="petal_length", data=df)
plt.title("Distribution de la longueur des pétales par espèce")
plt.show()
```

#Histogrammes


Utilisé pour visualiser la distribution d'une variable.


```python
plt.figure(figsize=(8,6))
sns.histplot(df["sepal_length"], kde=True)
plt.title("Distribution de sepal_length")
plt.show()
```

#Graphique à barres

 Utilisé pour comparer des moyennes entre catégories.

```python
plt.figure(figsize=(8,6))
sns.barplot(x="species", y="sepal_length", data=df)
plt.title("Longueur moyenne des sépales par espèce")
plt.show()
```

#Matrice de corrélation

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.drop('species', axis=1).corr(), annot=True, cmap="coolwarm")
plt.title("Matrice de corrélation")
plt.show()
```

#VISUALISATIONS AVEC PLOTLY (INTERACTIF)

Plotly permet des graphiques interactifs (zoom, info-bulles).

#Scatter interactif

```python
fig = px.scatter(df,
                 x="sepal_length",
                 y="sepal_width",
                 color="species",
                 title="Scatter Plot Interactif")
fig.show()
```

#Boxplot interactif

```python
fig = px.box(df,
             x="species",
             y="petal_length",
             color="species",
             title="Boxplot Interactif")
fig.show()
```

#Histogramme interactif

```python
fig = px.histogram(df,
                   x="sepal_length",
                   color="species",
                   marginal="box",
                   title="Histogramme Interactif")
fig.show()
```

#Scatter Matrix

```python
fig = px.scatter_matrix(df,
                        dimensions=["sepal_length",
                                    "sepal_width",
                                    "petal_length",
                                    "petal_width"],
                        color="species",
                        title="Matrice de dispersion")
fig.show()
```

*Les visualisations montrent que l’espèce Setosa est clairement séparée des deux autres, notamment grâce aux variables PetalLength et PetalWidth. Les espèces Versicolor et Virginica présentent un certain chevauchement.*

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler, MinMaxScaler

# Load the dataset using an existing file path
df = pd.read_csv('/content/sample_data/california_housing_train.csv')

# Drop the 'Unnamed: 0' index column if it exists
# Note: 'california_housing_train.csv' typically does not have an 'Unnamed: 0' column.
# This check is kept for robustness if a different dataset were used.
if 'Unnamed: 0' in df.columns:
    df = df.drop('Unnamed: 0', axis=1)

print(df.info())
```

#Outlier Removal

```python
import numpy as np

# Identify numerical columns for outlier removal
# Since all columns in california_housing_train.csv are numerical, we can select them all.
score_cols = df.select_dtypes(include=np.number).columns

for col in score_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Filtering the outliers
    df = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]

print(f"Dataset size after outlier removal: {df.shape}")
```



#Data Transformation (Scaling)



```python
# Scaling Numerical Features
scaler = StandardScaler() # Or MinMaxScaler()
df[score_cols] = scaler.fit_transform(df[score_cols])

print(df[score_cols].head())
```

#Final Cleaning Check

```python
print("Missing values per column:")
print(df.isnull().sum())

# Visualizing results
sns.boxplot(data=df[score_cols])
plt.title("Boxplot of Scaled Scores (Outliers Removed)")
plt.show()
```

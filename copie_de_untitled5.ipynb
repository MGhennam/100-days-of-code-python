#Exercise 2: List Operations

#Exercise 4: Set Operations

```python
# Create two sets
set1 = {1, 2, 3, 4, 5}
set2 = {4, 5, 6, 7, 8}

# Intersection
intersection = set1.intersection(set2)

# Union
union = set1.union(set2)

# Print results
print("Intersection:", intersection)
print("Union:", union)
```

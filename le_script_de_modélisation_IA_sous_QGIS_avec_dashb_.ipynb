### 1. Structure de la Base de Données Exemple (`.csv`)

Le fichier généré simule l'historique épidémiologique à l'échelle locale (ex: Communes/Districts). Chaque ligne associe des indicateurs climatiques extraits par télédétection à une variable de santé publique :

* **Variables Climatiques/Environnementales (Predictors) :**
* `Temp_Surface_LST` : Température de surface moyenne diurne (°C).
* `Indice_Veg_NDVI` : Densité de la végétation (Indice de -1 à 1).
* `Humidite_Rel` : Humidité relative de l'air (%).
* `Dist_Points_Eau_KM` : Proximité aux oueds, barrages ou eaux stagnantes.


* **Variable Sanitaire (Target) :**
* `Taux_Incidence_10k` : Nombre de cas enregistrés (ex: Leishmaniose ou Paludisme) pour 10 000 habitants.

### 2. Script de Modélisation IA (À exécuter dans la console Python de QGIS)

Ce script utilise la librairie `scikit-learn` (généralement intégrée dans l'environnement Python de QGIS) pour entraîner un modèle d'intelligence artificielle (**Random Forest Regressor**) et injecter automatiquement les prédictions du risque sanitaire dans la table attributaire de votre couche SIG.

````
# -*- coding: utf-8 -*-
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from qgis.core import QgsVectorLayer, QgsProject, QgsField
from PyQt5.QtCore import QVariant

# ==========================================
# 1. ENTRAÎNEMENT DE L'INTELLIGENCE ARTIFICIELLE
# ==========================================

# Remplacer par le chemin réel de votre fichier CSV généré
csv_path = "donnees_epidemiologie_climat_exemple.csv"
data = pd.read_csv(csv_path)

# Définition des critères environnementaux (X) et de la cible sanitaire (y)
X = data[['Temp_Surface_LST', 'Indice_Veg_NDVI', 'Humidite_Rel', 'Dist_Points_Eau_KM', 'Densite_Pop_KM2']]
y = data['Taux_Incidence_10k']

# Initialisation et entraînement de l'algorithme Random Forest
print("Entraînement du modèle d'IA en cours...")
model = RandomForestRegressor(n_estimators=150, random_state=42)
model.fit(X, y)

r2_score = model.score(X, y)
print(f"Modèle entraîné avec succès. Performance R² : {r2_score:.4f}")

# ==========================================
# 2. CARTOGRAPHIE DU RISQUE PRÉDIT SOUS QGIS
# ==========================================

# Nom exact de votre couche vectorielle (Polygones des communes/districts) chargée dans QGIS
layer_name = "Communes_Sante_Territoire"
layers = QgsProject.instance().mapLayersByName(layer_name)

if layers:
    layer = layers[0]
    layer.startEditing()

    # Création d'un champ de destination pour la prédiction de l'IA s'il n'existe pas
    field_name = 'Risq_Pred'
    if layer.fields().indexFromName(field_name) == -1:
        layer.addAttribute(QgsField(field_name, QVariant.Double, "Double", 10, 2))
        layer.updateFields()

    idx = layer.fields().indexFromName(field_name)

    # Parcours des entités géographiques pour appliquer l'IA
    count = 0
    for feature in layer.getFeatures():
        try:
            # Récupération des données spatiales actuelles de l'entité
            lst = float(feature['LST_climat'])
            ndvi = float(feature['NDVI_veg'])
            hum = float(feature['Humidite'])
            dist = float(feature['Dist_Eau'])
            pop = float(feature['Densite_Hab'])

            # Formatage pour le modèle de prédiction
            input_features = np.array([[lst, ndvi, hum, dist, pop]])

            # Calcul de l'indicateur de risque épidémiologique par l'IA
            predicted_incidence = model.predict(input_features)[0]

            # Injection de la valeur calculée dans la table QGIS
            layer.changeAttributeValue(feature.id(), idx, float(predicted_incidence))
            count += 1
        except Exception as e:
            # Gestion des valeurs nulles ou manquantes sur certaines communes
            continue

    layer.commitChanges()
    layer.triggerRepaint()
    print(f"Calcul terminé : {count} entités mises à jour avec l'indicateur 'Risq_Pred'.")
else:
    print(f"Erreur : La couche '{layer_name}' n'est pas ouverte dans votre projet QGIS.")
````

### 3. Guide de Construction du Dashboard Décisionnel dans QGIS

Pour transformer ces résultats complexes en un tableau de bord visuel interactif, suivez ces étapes de configuration :

#### A. Cartographie Thématique Graduée (Le fond cartographique)

1. Ouvrez les **Propriétés de la couche** `Communes_Sante_Territoire` $\rightarrow$ Onglet **Symbologie**.
2. Sélectionnez le mode de rendu **Gradué** et choisissez la colonne `Risq_Pred` (générée par l'IA).
3. Choisissez une rampe de couleur explicite (ex: `YlOrRd` - Jaune au Rouge foncé).
4. Cliquez sur **Classer** (mode *Ruptures naturelles (Jenks)*, 5 classes) pour isoler visuellement les foyers épidémiologiques prioritaires à l'échelle nationale.

#### B. Intégration des Graphiques Statistiques (Extension Data Plotly)

1. Installez l'extension **Data Plotly** depuis le gestionnaire d'extensions QGIS.

* *Axe X* : `Temp_Surface_LST`

#### C. Mise en Page Dynamique (Le Panneau de Bord Exécutif)

1. Ouvrez le **Gestionnaire de mise en page** de QGIS et créez une nouvelle mise en page nommée *Dashboard_Climat_Sante*.
2. Ajoutez votre carte principale.
3. Ajoutez les graphiques générés via *Data Plotly* (exportés en SVG/PNG).

Ce pipeline interconnecté permet aux décideurs d'ajuster les plans de riposte sanitaire (distribution de traitements, démoustication, renforcement des cliniques mobiles) directement à partir de prédictions environnementales anticipées par l'IA.
